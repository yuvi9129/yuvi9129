```python
import pandas as pd
import json
import nltk
import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud
from nltk.sentiment import SentimentIntensityAnalyzer

# Download necessary NLTK data
nltk.download('vader_lexicon')

# Load the JSON file
file_path = "D:\\3rd SEM\\message.json"
with open(file_path, encoding="utf-8") as f:
    data = json.load(f)

# Extract messages
messages = data["messages"]

# Convert JSON to DataFrame
chat_data = []
sia = SentimentIntensityAnalyzer()

for msg in messages:
    if "content" in msg:  # Only process text messages
        chat_data.append({
            "sender": msg["sender_name"],
            "message": msg["content"],
            "timestamp": pd.to_datetime(msg["timestamp_ms"], unit="ms"),
            "sentiment": sia.polarity_scores(msg["content"])["compound"]
        })

df = pd.DataFrame(chat_data)

# Categorize sentiment
df["sentiment_category"] = df["sentiment"].apply(lambda x: "Positive" if x > 0.05 else ("Negative" if x < -0.05 else "Neutral"))

# Aggregate message counts
message_counts = df["sender"].value_counts()

# Aggregate sentiment counts
sentiment_counts = df["sentiment_category"].value_counts()

# Generate word cloud
text = " ".join(df["message"].dropna())
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

# Plot messages per sender
plt.figure(figsize=(10, 5))
sns.barplot(x=message_counts.index, y=message_counts.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Messages per User")
plt.ylabel("Number of Messages")
plt.xlabel("User")
plt.show()

# Plot sentiment distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="coolwarm")
plt.title("Sentiment Distribution")
plt.ylabel("Number of Messages")
plt.xlabel("Sentiment Category")
plt.show()


# Plot daily message trend
plt.figure(figsize=(10, 5))
daily_messages.plot(kind="line", marker="o", color="blue")
plt.title("Messages Over Time")
plt.ylabel("Number of Messages")
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


# Plot hourly activity
plt.figure(figsize=(10, 5))
hourly_activity.plot(kind="bar", color="orange")
plt.title("Message Activity by Hour")
plt.ylabel("Number of Messages")
plt.xlabel("Hour of Day")
plt.xticks(rotation=0)
plt.grid(True)
plt.show()


# Display word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Messages")
plt.show()


# Display summary statistics
df.describe(include="all")
```
